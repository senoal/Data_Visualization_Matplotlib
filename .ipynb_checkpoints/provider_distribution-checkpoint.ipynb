# Visualize

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Provinsi

```python
def prov():
    provinsi = df1['provinsi'].value_counts().rename_axis('provinsi').reset_index(name='count')
    # print(provinsi,'\n')

    x = provinsi['provinsi']
    y = provinsi['count']
    # exp = y/30000

    plt.title("Provinsi")
    plt.barh(x, y, height=0.5)
    plt.show()
prov()
```

## Brand

```python
def brand():
    brand = df1['brand'].value_counts().rename_axis('brand').reset_index(name='count')
    # print(brand,'\n')

    brand = brand[brand['brand'] != 'UNKNOWN']
    # print(brand)

    x = brand['brand']
    y = brand['count']
    exp = y/30000

    plt.title("Brand")
    plt.pie(y, labels = x, startangle = 180,
            explode=exp, shadow = True,
           autopct = '%1.2f%%')
    plt.show()
brand()
```

## Jenis

```python
def jns():
    jenis = df1['jenis'].value_counts().rename_axis('jenis').reset_index(name='count')
#     print(jenis,'\n')

    x = jenis['jenis']
    y = jenis['count']
    exp = y/40000

    plt.title("Jenis")
    plt.pie(y, labels = x, startangle = 180,
            explode=exp, shadow = True,
           autopct = '%1.2f%%')
    plt.show()
jns()
```

## Customer Type (Pie)

```python
def cst():
    cust_type = df1['cust_type'].value_counts().rename_axis('cust_type').reset_index(name='count')
    # print(cust_type,'\n')

    x = cust_type['cust_type']
    y = cust_type['count']
    exp = y/80000

    plt.title("Type")
    plt.pie(y, labels = x, startangle = 180,
            explode=exp, shadow = True,
           autopct = '%1.2f%%')
    plt.show()
cst()
```

## Customer Type (Bar)

```python
def tp():
    cust_type = df1['cust_type'].value_counts().rename_axis('cust_type').reset_index(name='count')
#     print(cust_type,'\n')

    x_c = cust_type['cust_type']
    y_c = cust_type['count']
    # exp = y/80000

    plt.title("Type")
    plt.barh(x_c, y_c, height=0.8)
    for index, value in enumerate(y):
        plt.text(value, index, str(value))
    plt.show()
tp()
```
